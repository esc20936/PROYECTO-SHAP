# XGBoost

### - Setup

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import shap
```

```python
# Cargar los datos desde el archivo CSV
data = pd.read_csv("../data/external/Student_performance_data _.csv")
```

```python
# Definir la variable objetivo y los features
# Seleccionar las mismas que se usaron en el modelo Random Forest
target = data['GradeClass']
features = data[['StudyTimeWeekly', 'Absences', 'Tutoring', 'ParentalSupport', 'Extracurricular', 'GPA']]
```

```python
# Dividir los datos en entrenamiento y validación
# Esto permite evaluar el rendimiento del modelo en datos no vistos durante el entrenamiento
trainX, valX, trainY, valY = train_test_split(features, target, random_state=1)
```

```python
# Definir y entrenar el modelo XGBoost
# Usar XGBoost para construir un modelo de regresión
xgboostModel = xgb.XGBRegressor(random_state=1)
xgboostModel.fit(trainX, trainY)
```

```python
# Predecir en el conjunto de validación
predictions = xgboostModel.predict(valX)
```

```python
# Evaluar el modelo con MAE que nos dice en promedio cuánto se desvían las predicciones del valor real
print("El MAE del modelo XGBoost es: ")
print(mean_absolute_error(valY, predictions))
```

### - Análisis SHAP

```python
# Interpretación del modelo usando SHAP
# Ayuda a entender cómo cada característica influye en las predicciones del modelo
explainer = shap.Explainer(xgboostModel, trainX)
shapValues = explainer(valX)
```

```python
# Crear un gráfico de resumen SHAP para ver la importancia de las características
# Muestra cómo cada característica influye en las predicciones del modelo en general
shap.summary_plot(shapValues, valX)
```

```python
# Crear un gráfico de dependencia SHAP para analizar la relación entre una característica específica y las predicciones
# Muestra cómo los valores de la característica Absences afectan la predicción, teniendo en cuenta la interacción con GPA
shap.dependence_plot("Absences", shapValues.values, valX, interaction_index="GPA")
```

```python
# Crear un gráfico fuerza SHAP para analizar una predicción específica
# Desglosa cómo cada característica contribuye a una predicción individual
shap.force_plot(shapValues[0].base_values, shapValues[0].values, valX.iloc[0,:], matplotlib=True)
```

### - Conclusiones

### 1. SHAP Summary Plot

- GPA es la variable más influyente. Valores altos de GPA tienden a aumentar la predicción del modelo, mientras que valores bajos la disminuyen. Esto indica que el rendimiento académico GPA es el factor más determinante en las predicciones de GradeClass.
- StudyTimeWeekly y Absences son variables que tienen un gran impacto. El tiempo dedicado al estudio por semana afecta positivamente a las predicciones.
- ParentalSupport, Tutoring, y Extracurricular tienen un impacto menor.
- Este gráfico identifica qué características son más relevantes para las decisiones del modelo, áreas que se pueden mejorar si se lleva a la práctica el modelo con los datos.

### 2. SHAP Dependence Plot

- Mas ausencias reducen las predicciones del modelo. Esto es esperado porque faltar a clases afecta negativamente el rendimiento académico.
- Los estudiantes con un GPA bajo son muy afectados por las ausencias. Los de alto GPA son menos afectados, lo cual dice que los estudiantes con mejor rendimiento académico pueden manejar mejor la falta de asistencia.

### 3. SHAP Force Plot

- Absences, Tutoring y Extracurricular están empujando la predicción hacia abajo. GPA, ParentalSupport y StudyTimeWeekly tratan de aumentarla. Esto es esperado y refuerza la idea de que un buen rendimiento académico y apoyo parental son factores protectores que mejoran las predicciones.
