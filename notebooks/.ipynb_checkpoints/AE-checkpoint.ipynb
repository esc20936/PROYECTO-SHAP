# Análisis exploratorio

Importamos las librerias que usaremos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Cargaremos nuestros datos

```python
students_data = pd.read_csv('../data/external/Student_performance_data _.csv')
```

Queremos saber que columnas tiene nuestro data set

```python
students_data.columns
```

```python
students_data.head()
```

Queremos conocer algunos datos estadisticos de nuestro dataset

```python
students_data.describe()
```

Queremos ver si existen valores nulos

```python
# look for missing values
students_data.isnull().sum()
```

Nuestro dataset esta limpio

Queremos saber la correlación de nuestras variables

```python
correlation_matrix = students_data.corr()
```

```python
sns.heatmap(correlation_matrix, cmap='coolwarm', vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Quitaremos el grade class y el studentId ya que no son necesarias para nuestro modelo

```python
students_data_cleaned = students_data.drop(['GradeClass', 'StudentID'], axis=1)
```

```python
correlation_matrix = students_data_cleaned.corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Podemos observar que nuestros features más importantes son studyTimeWeekly, Absences, Tutoring, ParentalSupport,GPA y ExtraCurricular
